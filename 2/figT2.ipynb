```python
import json
import matplotlib.pyplot as plt

proc_num = [2000, 5000, 10000, 20000, 30000, 40000, 50000]
data = 'Error'
with open('res2.txt', 'r') as f:
    data = json.load(f)
x = range(1,len(data[0])+1)
plt.figure(figsize=(12,8))
plt.xlabel("Number of processes, $p$", fontsize=24)
plt.ylabel("Speedup, $N$", fontsize=24)
plt.xticks(x)
for (y, n) in zip(data, proc_num):
    plt.plot(x, y, label=str(n))
plt.legend(fontsize=16)
plt.show()
```
