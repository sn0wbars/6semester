```python
import json
import numpy as np
import matplotlib.pyplot as plt

with open('res3.txt', 'r') as f:
    data = json.load(f)
    x = range(1,len(data)+1)
    plt.figure(figsize=(12,8))
    plt.xlabel("Number of processes, $p$", fontsize=24)
    plt.ylabel("Speedup, $N$", fontsize=24)
    plt.xticks(x)
    plt.plot(x, data[0]/np.array(data), label='10^9')
    plt.plot(x,x)
    plt.legend(fontsize=16)
    plt.show()
```
